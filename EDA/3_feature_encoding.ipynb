```python
import pandas as pd
import numpy as np
import os
import seaborn as sns
import matplotlib.pyplot as plt
from enum import Enum
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('./processed/feature_binned.csv')
df
```

```python
encoding_dic = {
    'Yes' : 1,
    'No' : 0
}

yes_no_columns = ['Partner', 'Dependents' ,'PhoneService', 'PaperlessBilling', 'Churn' ]

for col in yes_no_columns:
    df[col] =df[col].map(encoding_dic)
```

```python
#Ordinal encoding for tenure

tenure_dic = {
    'new':1,
    'loyal': 2,
    'established':3
}

df['tenure_bins'] = df['tenure_bins'].map(tenure_dic)
```

```python
categorical_columns= ['gender', 'MultipleLines', 'InternetService', 'OnlineSecurity', 'OnlineBackup', 'DeviceProtection', 'TechSupport', 'StreamingTV', 'StreamingMovies', 'Contract', 'PaymentMethod' ]
```

```python
for col in categorical_columns:
    column_dummies = pd.get_dummies(df[col], prefix=col)
    df = pd.concat([df,column_dummies], axis=1)
    del df[col]
```
