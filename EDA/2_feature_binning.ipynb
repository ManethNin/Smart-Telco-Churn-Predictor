```python
def tenureIntoBins(data):
    if (0 <= data <13):
        return 'new'
    elif (13 <= data < 49):
        return 'established'
    elif (49 <= data):
        return 'loyal'
    
df['tenure_bins'] = df['tenure'].apply(tenureIntoBins)
df
```
